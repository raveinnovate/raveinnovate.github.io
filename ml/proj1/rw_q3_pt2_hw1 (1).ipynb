```python
# Import Statements

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error
import seaborn as sns
```

```python
# PART A: Generate Dataset (LOW RANGE, training)
X_train= np.linspace(0, 5, num=20)
X_train = X_train.reshape(-1, 1)
print("X:")
print(X_train)


noise = np.random.normal(0, 1, size=X_train.shape)  # small random noise
y_train = 3 * X_train + 7 + noise
print("y")
print(y_train)
```

```python
# PART A: Generate Dataset (HIGH RANGE, test)
X_test= np.linspace(10, 15, num=20)
X_test = X_test.reshape(-1, 1)
print("X:")
print(X_test)

# No need to generate y, cause that is what we are predicting.
```

```python
print("Training Model...")
model1 = LinearRegression()
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)
print("Model trained.")
print("-" * 30)
```

```python
plt.scatter(X_train, y_train, color='blue', label='Training Data (0-5)')
plt.scatter(X_test, y_pred, color='red', label='Test Data (10-15)')

x_line = np.linspace(0, 15, 100).reshape(-1, 1)
y_line = model1.predict(x_line)
plt.plot(x_line, y_line, color='green', linewidth=2, label='Regression Line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression with Training and Test Data")
plt.legend()
plt.show()
```
